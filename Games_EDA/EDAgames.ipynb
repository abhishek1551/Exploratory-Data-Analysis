"Let's now apply some machine learning algorithms and evaluate their performance. To do this, first, we will need to convert categorical features in to numerical one. We can use label Encoder for this."


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data = df.copy()
```

```python
feature = ["Platform", "Genre"]


for col in feature:
    data[col] = le.fit_transform(df[col])
```

```python
#Split feature and target variables
X = data[['Platform', 'Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].values

y = data['Global_Sales'].values
```

```python
#Splitting the data into train and test ser
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=45)
```

We will now apply linear regression, Random forest regressor and AdaBoost models and will also be performing hyperparameter tuning for the best results.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, r2_score
```

```python
# Initialize Linear Regressor 
modellr = LinearRegression()
modellr.fit(X_train,y_train)
```

```python
pred= modellr.predict(X_test)
r2_lr =r2_score(y_test, pred)
```

```python
print(r2_lr)
```

Implementing Random forest regressor model.

```python
from sklearn.ensemble import RandomForestRegressor
# Initialize ARandom forest Regressor with default parameters
rfr = RandomForestRegressor(random_state= 40)
```

```python
rfr.fit(X_train, y_train)

pred = rfr.predict(X_test)

print(r2_score(y_test, pred))
```

Implementing AdaBoost regressor model.

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python

# Initialize AdaBoost Regressor with default parameters
ada_regressor = AdaBoostRegressor(random_state= 35)
```

```python
ada_regressor.fit(X_train, y_train)
y_pred = ada_regressor.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
print(f'R2 Score: {r2}')
```

Now, let's apply hyperparameter tuning for each model

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Initialize Random Forest Regressor
rfr = RandomForestRegressor(random_state=40)
```

```python
# Define hyperparameters to tune
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rfr, param_grid=param_grid, cv=3, scoring='r2', n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print(f'Best Parameters: {best_params}')
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
```

```python
#Evaluate
r2 = r2_score(y_test, y_pred)
print(f'R2 Score: {r2}')
```

```python
#AdaBoost
# Define hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}

```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=ada_regressor, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
# Best parameters found by GridSearchCV
best_params = grid_search.best_params_
print(f'Best Parameters: {best_params}')
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
```

```python
#Evaluate
r2 = r2_score(y_test, y_pred)
print(f'R2 Score: {r2}')
```

In conclusion, Hyperparameter tuning helped to increase the performance of the models significantly.
